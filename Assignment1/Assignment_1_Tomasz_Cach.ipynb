## Exercise 4
"I am interested in the house with 3 bedrooms, 2 bathrooms and 2 floors"  
a) Compute the basic statistics of apartments fulfilling above constraints.  
b) Distinct accepted apartments on any of previously created plots.  
c) Find such apartments with best price to footage ratio.  

```python
#a)
aps = data.loc[data['bedrooms'] == 3].loc[data['bathrooms'] == 2].loc[data['floors'] == 2]
display(aps)

# Basic statistics
display(aps.describe())
```

```python
#b)
fig, ax = plt.subplots(figsize=[14,8])

ax.scatter(x=data['long'], y=data['lat'], label="Every House")
ax.scatter(x=aps['long'], y=aps['lat'], label="Selected apartments")
ax.legend()

plt.show()
```

```python
#c)
# new column
aps['price_footage_ratio'] = aps['price'] / aps['sqft_living']

aps = aps.sort_values("price_footage_ratio", ascending=True).head(10)
display(aps)
```
